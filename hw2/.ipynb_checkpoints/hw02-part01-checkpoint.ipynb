```python
import pandas as pd
import glob
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg' 
```

```python
#read files and create dataframe names
read_files = glob.glob("./babynames/*.txt")
df_list = []
for file in glob.glob('./babynames/*.txt'):
    # Tweak this to work for your actual filepaths, if needed.
    year = int(file[15:19])
    names = (pd.read_csv(file,header=None)
          .assign(Year=year))
    df_list.append(names)

names = pd.concat(df_list)
names.columns = ['Name', 'Gender', 'Quantity', 'Year']
```

# Популярные женские имена в 10 интервалах

```python
years = list(range(1880,2011,1))
split_years = np.array_split(years,10)
female_split = np.array_split(female, 10)
female_group10 = []
popular_female_name = []
df = pd.DataFrame()

for i in range(len(split_years)):
    for year in split_years[i]:
        df = df.append(female[female['Year']==year])
    female_group10.append(df)
    df = pd.DataFrame()

female_group10_agg = []
for i in range(len(female_group10)):
    female_group10_agg.append(female_group10[i].groupby('Name').aggregate("sum"))

for i in range(len(female_group10_agg)):
    popular_female_name.append(female_group10_agg[i]['Quantity'].idxmax())
```

```python
print(popular_female_name)
```

# Популярные мужские имена в 10 интервалах

```python
male_split = np.array_split(male, 10)
male_split = np.array_split(male, 10)
male_group10 = []
popular_male_name = []
df = pd.DataFrame()

for i in range(len(split_years)):
    for year in split_years[i]:
        df = df.append(male[male['Year']==year])
    male_group10.append(df)
    df = pd.DataFrame()

male_group10_agg = []
for i in range(len(male_group10)):
    male_group10_agg.append(male_group10[i].groupby('Name').aggregate("sum"))

for i in range(len(male_group10_agg)):
    popular_male_name.append(male_group10_agg[i]['Quantity'].idxmax())
```

```python
print(popular_male_name)
```

# Количество имен, которое покрывает 50% людей.

```python
names.sort_values(by=['Year','Quantity'], inplace = True, ascending=False)
names_grouped_year = names.groupby('Year')

names_median = []

for year in years:
    names_median.append(pd.Index(names_grouped_year.get_group(year).Quantity.cumsum()).get_loc(
    names_grouped_year.get_group(year)['Quantity'].sum()/2, 'backfill'))

```

```python
print(names_median)
```

```python
plt.plot(years, names_median)
plt.ylabel('#Names median')
plt.xlabel('Year')
plt.show()
```

# Распределение по первой и последней букве имени для 1958, 1962, 1988 и 2005 года

```python
sel_df = names[(names['Year']==1958)| (names['Year']==1962) | (names['Year']==1988) | (names['Year']==2005)]

def first_letter (x):
    return x[0]

def last_letter (x):
    return x[-1]

sel_df['first_letter'] = sel_df['Name'].apply(first_letter)
sel_df['last_letter']  = sel_df['Name'].apply(last_letter)
```

# Распределение по первой букве имени за 4 года

```python
sns.catplot(x="first_letter", kind="count", data=sel_df, hue="Year", 
            order = sel_df['first_letter'].value_counts().index);
```

# Распределение по последней букве имени за 4 года

```python
sns.catplot(x="last_letter", kind="count", data=sel_df, hue="Year", 
            order = sel_df['last_letter'].value_counts().index);
```

# Как известные люди влияли на динамику имен

```python
#dataset of most famous people in history
pantheon = pd.read_csv("pantheon.tsv", sep='\t')
pantheon_us = pantheon[pantheon["countryCode"]=='US']
pantheon = pantheon.dropna(subset=['birthyear'])

def str_to_int (x):
    return int(x)

pantheon_us['birthyear_int'] = pantheon_us['birthyear'].apply(str_to_int)
# Historical Popularity Index - HPI 

pantheon_us_1900 = pantheon_us[pantheon_us["birthyear_int"] >= 1900]

top5_us_person_names = []
top5_us_person_HPI =pantheon_us_1900.nlargest(5, 'HPI')
for i in range(0,5,1):
    top5_us_person_names.append( ( (top5_us_person_HPI.iloc[i,1]).split(' ',1)[0], top5_us_person_HPI.iloc[i,23]) ) 
    
print(top5_us_person_names)
#some_string.split(' ', 1)[0]
```

```python
f, axes = plt.subplots(2,2)
#plt.subplots_adjust(bottom=0.1, right=0.8, top=0.9)
p1 = sns.lineplot(x="Year", y="Quantity", 
            data=names[names['Name']==top5_us_person_names[0][0]],ci=None, ax = axes[0][0])
p1.set_title(top5_us_person_names[0][0])
p1.axvline(x=top5_us_person_names[0][1], ymin=0, ymax=1,color='r')

p2 = sns.lineplot(x="Year", y="Quantity", 
            data=names[names['Name']==top5_us_person_names[1][0]],ci=None, ax = axes[0][1])
p2.set_title(top5_us_person_names[1][0])
p2.axvline(x=top5_us_person_names[1][1], ymin=0, ymax=1,color='r')

p3 = sns.lineplot(x="Year", y="Quantity", 
            data=names[names['Name']==top5_us_person_names[2][0]],ci=None, ax = axes[1][0])
p3.set_title(top5_us_person_names[2][0])
p3.axvline(x=top5_us_person_names[2][1], ymin=0, ymax=1,color='r')

p4 = sns.lineplot(x="Year", y="Quantity", 
            data=names[names['Name']==top5_us_person_names[3][0]],ci=None, ax = axes[1][1])
p4.set_title(top5_us_person_names[3][0])
p4.axvline(x=top5_us_person_names[3][1], ymin=0, ymax=1,color='r')
```
